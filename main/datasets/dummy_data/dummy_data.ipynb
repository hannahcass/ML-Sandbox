```python
import pandas as pd
import numpy as np
from faker import Faker
import random
from datetime import datetime, timedelta

fake = Faker()
```

```python
manufacturers_models = {
    'Apple': ['iPhone 11', 'iPhone 12', 'iPhone 13', 'iPhone 14', 'iPhone 15', 'iPhone 16'],
    'Samsung': ['Galaxy S10', 'Galaxy S20', 'Galaxy S21', 'Galaxy S22', 'Galaxy S23', 'Galaxy S24'],
    'Google': ['Pixel 4', 'Pixel 5', 'Pixel 6', 'Pixel 7', 'Pixel 8', 'Pixel 9'],
    'Huawei': ['P30', 'P40', 'P50', 'P60', 'Mate 30', 'Mate 40', 'Mate 50', 'Mate 60'],
    'OnePlus': ['OnePlus 7', 'OnePlus 8', 'OnePlus 9', 'OnePlus 10', 'OnePlus 11', 'OnePlus 12'],
    'Xiaomi': ['Mi 9', 'Mi 10', 'Mi 11', 'Mi 12', 'Mi 13', 'Mi 14'],
    'Sony': ['Xperia 1', 'Xperia 5', 'Xperia 10', 'Xperia 15', 'Xperia 20'],
    'LG': ['G8', 'G9', 'V50', 'V60', 'V70'],
    'Nokia': ['8.1', '9 PureView', '7.2', 'X20', 'X30'],
    'Motorola': ['Moto G7', 'Moto G8', 'Moto G9', 'Moto G10', 'Moto G11', 'Moto G12']
}

release_years = {
    'iPhone 11': 2019, 'iPhone 12': 2020, 'iPhone 13': 2021, 'iPhone 14': 2022, 'iPhone 15': 2023, 'iPhone 16': 2024,
    'Galaxy S10': 2019, 'Galaxy S20': 2020, 'Galaxy S21': 2021, 'Galaxy S22': 2022, 'Galaxy S23': 2023, 'Galaxy S24': 2024,
    'Pixel 4': 2019, 'Pixel 5': 2020, 'Pixel 6': 2021, 'Pixel 7': 2022, 'Pixel 8': 2023, 'Pixel 9': 2024,
    'P30': 2019, 'P40': 2020, 'P50': 2021, 'P60': 2023, 'Mate 30': 2019, 'Mate 40': 2020, 'Mate 50': 2022, 'Mate 60': 2024,
    'OnePlus 7': 2019, 'OnePlus 8': 2020, 'OnePlus 9': 2021, 'OnePlus 10': 2022, 'OnePlus 11': 2023, 'OnePlus 12': 2024,
    'Mi 9': 2019, 'Mi 10': 2020, 'Mi 11': 2021, 'Mi 12': 2022, 'Mi 13': 2023, 'Mi 14': 2024,
    'Xperia 1': 2019, 'Xperia 5': 2020, 'Xperia 10': 2021, 'Xperia 15': 2023, 'Xperia 20': 2024,
    'G8': 2019, 'G9': 2020, 'V50': 2019, 'V60': 2020, 'V70': 2023,
    '8.1': 2019, '9 PureView': 2019, '7.2': 2020, 'X20': 2023, 'X30': 2024,
    'Moto G7': 2019, 'Moto G8': 2020, 'Moto G9': 2021, 'Moto G10': 2022, 'Moto G11': 2023, 'Moto G12': 2024
}

height_range = (5, 7)
width_range = (2, 3)
weight_range = (140, 240)

camera_resolutions = [8, 12, 16, 20]
memory_sizes = [16, 32, 64, 128, 256, 512]
num_sim_cards_options = [1, 2]
gps_options = [0, 1]
touchscreen_options = [0, 1]
colors = ['Red', 'Green', 'Blue', 'Black', 'White', 'Pink', 'Purple', 'Yellow', 'Orange', 'Gray']

phones = []
for manufacturer, model_list in manufacturers_models.items():
    for model in model_list:
        for _ in range(10):
            phone = {
                'item_id': fake.random_int(min=1000, max=9999),
                'item_name': model,
                'item_release_year': release_years[model],
                'manufacturer': manufacturer,
                'camera': random.choice(camera_resolutions),
                'height': random.randint(*height_range),
                'width': random.randint(*width_range),
                'weight': random.randint(*weight_range),
                'num_sim_cards': random.choice(num_sim_cards_options),
                'gps': random.choice(gps_options),
                'touchscreen': random.choice(touchscreen_options),
                'color': random.choice(colors),
                'memory': random.choice(memory_sizes)
            }
            phones.append(phone)

while len(phones) < 300:
    manufacturer = random.choice(list(manufacturers_models.keys()))
    model = f"{random.choice(manufacturers_models[manufacturer])} {fake.random_int(min=1, max=10)}"
    phone = {
        'item_id': fake.random_int(min=1000, max=9999),
        'item_name': model,
        'item_release_year': random.choice([2023, 2024]),
        'manufacturer': manufacturer,
        'camera': random.choice(camera_resolutions),
        'height': random.randint(*height_range),
        'width': random.randint(*width_range),
        'weight': random.randint(*weight_range),
        'num_sim_cards': random.choice(num_sim_cards_options),
        'gps': random.choice(gps_options),
        'touchscreen': random.choice(touchscreen_options),
        'color': random.choice(colors),
        'memory': random.choice(memory_sizes)
    }
    phones.append(phone)


phones_df = pd.DataFrame(phones)
```

```python
def get_better_phone(current_phone):
    better_phones = phones_df[(phones_df['item_release_year'] >= current_phone['item_release_year']) &
                              (phones_df['memory'] >= current_phone['memory']) &
                              (phones_df['camera'] >= current_phone['camera'])]
    if better_phones.empty:
        return current_phone
    return better_phones.sample(1).iloc[0]
```

```python
customers = []

for customer_id in range(1, 50001):
    first_phone = phones_df.sample(1).iloc[0]
    contract_start_date = datetime(2018, 1, 1) + timedelta(days=random.randint(0, 365))
    contract_duration_days = random.randint(365, 730)
    contract_end_date = contract_start_date + timedelta(days=contract_duration_days)
    
    customers.append({
        'customer_id': customer_id,
        'item_id': first_phone['item_id'],
        'contract_start_date': contract_start_date,
        'contract_end_date': contract_end_date
    })
    
    contract_start_date = contract_end_date + timedelta(days=random.randint(30, 90))
    second_phone = get_better_phone(first_phone)
    contract_duration_days = random.randint(365, 730)
    contract_end_date = contract_start_date + timedelta(days=contract_duration_days)
    
    customers.append({
        'customer_id': customer_id,
        'item_id': second_phone['item_id'],
        'contract_start_date': contract_start_date,
        'contract_end_date': contract_end_date
    })
    
    if random.random() < 0.5:
        contract_start_date = contract_end_date + timedelta(days=random.randint(30, 90)) 
        third_phone = get_better_phone(second_phone)
        contract_duration_days = random.randint(365, 730)
        contract_end_date = contract_start_date + timedelta(days=contract_duration_days)
        
        customers.append({
            'customer_id': customer_id,
            'item_id': third_phone['item_id'],
            'contract_start_date': contract_start_date,
            'contract_end_date': contract_end_date
        })


customers_df = pd.DataFrame(customers)


result_df = pd.merge(customers_df, phones_df, on='item_id', how='left')

```

```python
result_df.to_csv('customer_phone_contracts_2.csv', index=False)
```
